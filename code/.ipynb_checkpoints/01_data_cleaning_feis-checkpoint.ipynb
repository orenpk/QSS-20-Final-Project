```python
## Import packages
import pandas as pd
```

```python
## Load data
sirs_demo = pd.read_pickle('../data/sirs_demo_abc_clean')
feis = pd.read_excel('../data/Dartmouth FEIS Data.xlsx')
```

```python
sirs_demo
```

```python
feis
```

```python
# Remove duplicate Local ID values from the SIRS dataset
sirs_demo_unique = sirs_demo.drop_duplicates(subset='Local ID')

feis_cleaned = feis.dropna(subset=["What\nadvice would you give to service planners regarding the mental health service\nneeds of persons with IDD and their families?"])
```

```python
print(sirs_demo_unique.shape)
print(sirs_demo.shape)
print(feis_cleaned.shape)
print(feis.shape)
```

```python
# Renaming the column in FEIS dataset to match the SIRS dataset for merging
feis_cleaned.rename(columns={'Respondent ID #  (SIRS Local ID)': 'Local ID'}, inplace=True)

# Merging the datasets on 'Local ID'
merged_data = pd.merge(sirs_demo_unique, feis_cleaned, on='Local ID', how='inner')
```

```python
print(sirs_demo_unique.shape)
print(feis_cleaned.shape)
print(merged_data.shape)
```

```python
merged_data.to_pickle('../data/feis_sirs_demo_merged')
```
